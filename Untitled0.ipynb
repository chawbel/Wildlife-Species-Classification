```python
import pandas as pd

# Load OpenMV labels
openmv_df = pd.read_csv("camera_trap_images/6mhrhn7rxc-1/OpenMV_labels.csv")
print(openmv_df.head())

# Load RaspberryPi labels
raspberrypi_df = pd.read_csv("camera_trap_images/6mhrhn7rxc-1/RaspberryPi_labels.csv")
print(raspberrypi_df.head())
```

```python
openmv_df["source"] = 'OpenMV'
raspberrypi_df["source"] = "RaspberryPi"

#combine the dataframes
combined_df = pd.concat([openmv_df,raspberrypi_df], ignore_index=True)

combined_df.head()
```

```python
combined_df.info()
```

```python
#cleaning the species column
combined_df['SPECIES'] = combined_df['SPECIES'].str.strip('[]')
```

```python
combined_df = combined_df[['FILENAME','SPECIES','source']]

# Split multi-species entries into separate rows
combined_df = combined_df.assign(SPECIES=combined_df['SPECIES'].str.split(',')).explode('SPECIES')
combined_df['SPECIES'] = combined_df['SPECIES'].str.strip()  # Remove leading/trailing spaces

# Save the cleaned CSV for reference
combined_df.to_csv("cleaned_labels.csv", index=False)

combined_df.head()
```

```python
import os
import shutil

# Organize images into species folders
for _, row in combined_df.iterrows():
    species = row["SPECIES"]  # Species label
    source = row["source"]    # Source (OpenMV or RaspberryPi)
    filename = row["FILENAME"]  # Image filename

    if source == "OpenMV":
        src = f"camera_trap_images/6mhrhn7rxc-1/OpenMV_images/{filename}"
    elif source == "RaspberryPi":
        found = False
        for i in range(1, 31):  # Loop through 1st-deployment to 30th-deployment
            deployment_folder = f"{i}{'st' if i == 1 else ('nd' if i == 2 else ('rd' if i == 3 else 'th'))}-deployment"
            src = f"camera_trap_images/6mhrhn7rxc-1/RaspberryPi_images/{deployment_folder}/{filename}"
            print(f"Checking path: {src}")  # Debugging output
            if os.path.exists(src):
                found = True
                break  # Stop searching once the file is found

        if not found:
            print(f"File not found: RaspberryPi_images/*/{filename}")
            continue

    # Destination path
    dst_dir = f"dataset/train/{species}"
    os.makedirs(dst_dir, exist_ok=True)
    dst = f"{dst_dir}/{filename}"

    # Copy the image
    try:
        shutil.copy(src, dst)
        print(f"Copied: {src} -> {dst}")
    except Exception as e:
        print(f"Error copying {src}: {e}")
```

```python
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import torch
# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),       # Resize images to 224x224 (ResNet input size)
    transforms.RandomHorizontalFlip(),  # Randomly flip images for augmentation
    transforms.ToTensor(),              # Convert images to tensors
    transforms.Normalize(               # Normalize pixel values
        mean=[0.485, 0.456, 0.406],      # Pretrained model's normalization stats
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
train_dataset = ImageFolder("dataset/train", transform=transform)
```

```python
# Split into train and validation sets
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])
```

```python
# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
import torch.nn as nn
import torch.optim as optim
from torchvision import models

# Load pretrained ResNet50
model = models.resnet50(pretrained=True)

# Freeze all layers except the final layer
for param in model.parameters():
    param.requires_grad = False

# Replace the final fully connected layer
num_classes = len(train_dataset.dataset.classes)  # Number of species
model.fc = nn.Linear(model.fc.in_features, num_classes)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True  # Ignore truncated images

# Custom Dataset Class (to handle errors)
class SafeImageFolder(ImageFolder):
    def __getitem__(self, index):
        try:
            return super().__getitem__(index)
        except Exception:
            # Return a dummy image and label if there's an error
            return (torch.zeros(3, 224, 224), 0)  # Adjust dimensions if needed

# Load dataset with error handling
train_dataset = SafeImageFolder("dataset/train", transform=transform)
```

```python
# Reload datasets with the custom class
train_dataset = SafeImageFolder("dataset/train", transform=transform)
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Train the model
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Skip dummy images (if label is 0 and image is all zeros)
        if torch.all(images == 0):
            continue

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}")

print("Training complete!")
```

```python
# Evaluate on validation set
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Validation Accuracy: {100 * correct / total:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate predictions
y_true = []
y_pred = []
model.eval()
with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

# Plot confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=train_dataset.dataset.classes, yticklabels=train_dataset.dataset.classes)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Generate predictions and confidence scores
val_predictions = []
val_true_labels = []
val_confidences = []

model.eval()
with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        probabilities = torch.softmax(outputs, dim=1)  # Compute probabilities for the entire batch
        confidence, predicted = torch.max(probabilities, 1)  # Get max confidence and predicted class

        val_predictions.extend(predicted.cpu().numpy())
        val_true_labels.extend(labels.cpu().numpy())
        val_confidences.extend(confidence.cpu().numpy())  # Convert to numpy and store confidence
```

```python
# Create a DataFrame for Power BI
val_df = pd.DataFrame({
    "True Label": [train_dataset.dataset.classes[l] for l in val_true_labels],
    "Predicted Label": [train_dataset.dataset.classes[p] for p in val_predictions],
    "Confidence": [round(c * 100, 2) for c in val_confidences]  # Convert to percentage
})

val_df.to_csv("validation_predictions.csv", index=False)
```

```python
import gradio as gr

# Load the trained model
model.load_state_dict(torch.load("animal_classifier.pth"))
model.eval()

# Define the prediction function
def predict_species(image):
    image = transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        outputs = model(image)
        probabilities = torch.softmax(outputs, dim=1)[0]
        predicted_class = torch.argmax(probabilities).item()
        species = train_dataset.dataset.classes[predicted_class]
        confidence = probabilities[predicted_class].item() * 100
    return {species: confidence for species, confidence in zip(train_dataset.dataset.classes, probabilities)}

# Create a Gradio interface
iface = gr.Interface(
    fn=predict_species,
    inputs=gr.Image(label="Upload a Camera Trap Image"),
    outputs=gr.Label(num_top_classes=3),
    examples=["dataset/train/IMPALA/snapshot_201910831124745.jpg"]  # Add example images
)

# Launch the app
iface.launch()
```
